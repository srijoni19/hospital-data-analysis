```python
#load required libraries and dataset

import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('hospital_data.csv')
```

```python
#task 1 - disease distribution (pie chart)

#count patients per disease
diease_counts = df['disease'].value_counts()

#plot pie chart
plt.figure(figsize=(8,8))

plt.pie(diease_counts, labels=diease_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Disease Distribution of Patients")
plt.axis('equal')
plt.show()
```

```python
#task 2 - doctor workload (bar chart)

#count patients per doctor
doctor_counts = df['doctor'].value_counts()

#plot bar chart
plt.figure(figsize=(10,6))
plt.bar(doctor_counts.index, doctor_counts.values, color='skyblue')
plt.title("Number of Patients per Doctor")
plt.xlabel("Doctor")
plt.ylabel("Number of Patients")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#extra - patients vs avg cost per disease (scatter plot)

disease_stats = df.groupby('disease').agg({
    'patient_id': 'count',
    'treatment_cost': 'mean'
}).reset_index()

plt.figure(figsize=(10,6))
plt.scatter(disease_stats['patient_id'], disease_stats['treatment_cost'], color='orange')
plt.title("Patients vs Average Treatment Cost per Disease")
plt.xlabel("Number of Patients")
plt.ylabel("Average Treatment Cost")

    
plt.tight_layout()  
plt.show()
```
